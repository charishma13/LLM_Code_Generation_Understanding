```python
numeric_columns = list(set(modified_data.columns.tolist()) ^ set(['Cell Type',
       'Cell Side Lengths [a, b, c] (Å)', 'Cell Angles [α, β, γ] (degrees)']))
```

```python
modified_data[numeric_columns] = modified_data[numeric_columns].apply(lambda x: pd.to_numeric(x.str.strip()))  # This will convert strings to numbers, invalid ones to NaN
```

```python
modified_data = modified_data.apply(lambda x: x.str.replace('\xa0', ' ') if x.dtype == 'object' else x)
modified_data
```
